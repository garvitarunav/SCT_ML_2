## Importing the DATA and essential libraries

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, adjusted_rand_score, adjusted_mutual_info_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

#### Segregating the Target Variable from the Input Features

```python
target_var = data['CustomerID']
```

```python
feature_var = data.drop('CustomerID', axis = 1)
```

```python
feature_var.head()
```

```python
feature_var['Gender'] = feature_var['Gender'].replace('Male',0)
feature_var['Gender'] = feature_var['Gender'].replace('Female',1)
```

```python
feature_var.head()
```

#### Extracting the main input features and standardizing them using the standard scaler 

```python
x = feature_var[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

### Using the Elbow Method for checking the optimal number of clusters

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

#### Visualizing the above for getting the best number of clusters

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Training the data using the KMeans Algorithm

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(x_scaled)
```

```python
data['Cluster'] = y_kmeans
```

#### Visualizing the Clusters 

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data, palette='viridis', s=100, alpha=0.7)
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### Model Evaluation

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(x_scaled)
```

```python
inertia = kmeans.inertia_
print(f"Inertia (WCSS): {inertia}")
```

```python
silhouette_avg = silhouette_score(x_scaled, y_kmeans)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
davies_bouldin = davies_bouldin_score(x_scaled, y_kmeans)
print(f"Davies-Bouldin Index: {davies_bouldin}")
```

```python
calinski_harabasz = calinski_harabasz_score(x_scaled, y_kmeans)
print(f"Calinski-Harabasz Index: {calinski_harabasz}")
```
